```python
import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.parse("data/data.xml")
root = tree.getroot()

rows = []
for child in root:
    row = {elem.tag: elem.text for elem in child}
    rows.append(row)

df = pd.DataFrame(rows)
print(df)
```
